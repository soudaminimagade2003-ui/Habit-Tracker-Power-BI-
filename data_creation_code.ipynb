## Import Libraries

```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

## Create Users File

```python
users = []

genders = ["Male", "Female"]

goals = [
    "Overall Health Improvement",
    "Financial Growth",
    "Mental Wellbeing",
    "Productivity Improvement",
    "Balanced Lifestyle",
    "Self Development",
    "Discipline & Consistency",
    "Stress Reduction"
]

join_date = datetime(2026, 1, 1)

for i in range(1, 201):

    age = random.randint(18, 50)
    gender = random.choice(genders)
    goal = random.choice(goals)

    users.append([
        i,
        "User_" + str(i),
        age,
        gender,
        join_date.date(),
        goal
    ])

users_df = pd.DataFrame(users, columns=[
    "user_id",
    "user_name",
    "age",
    "gender",
    "join_date",
    "goal"
])

users_df.to_csv("users.csv", index=False)

print("Users file created successfully")
```

## Create Habits File

```python
habits = [
    ("Walk", "Fitness"),
    ("Gym Workout", "Fitness"),
    ("Yoga", "Fitness"),
    ("Savings", "Finance"),
    ("Meditation", "Mental Health"),
    ("Journaling", "Mental Health"),
    ("Hydration", "Wellness"),
    ("Nutrition", "Wellness"),
    ("Diet", "Wellness"),
    ("Sleep", "Wellness"),
    ("No Junk Food", "Wellness"),
    ("Skincare", "Wellness"),
    ("Reduce Screen Time", "Wellness"),
    ("Hobbies", "Growth"),
    ("Core Work", "Productivity")
]

habits_df = pd.DataFrame(habits, columns=["habit_name", "category"])
habits_df.to_csv("habits.csv", index=False)

print("Habits file created")
```
